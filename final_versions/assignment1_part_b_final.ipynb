$\small \underline{\textbf{Question - 1}} \quad \text{Evaluate the following expression:}$
$ \small (7 \,\textasciicircum\, 3) * (8 \gg 2) + 9 \,\%\, 4 $

$$
\small \text{Breaking down the expression into blocks and operating on these individual blocks in tabular form}
$$

$$
\begin{array}{clccc}
\hline
\text{Block} & \text{Operation} & \text{Decimal A} & \text{Decimal B} & \text{Result} \\
\hline
\text{A} & \text{XOR ($\textasciicircum$)} & 7 & 3 & 
\begin{array}{c}
0111 \\
0011 \\ \hline
0100 \; \Rightarrow \; 4
\end{array} \\
\hline
\text{B} & \text{Right \,Shift ($\gg$)} & 8 & 2 &
\begin{array}{c}
1000 \\
\gg 2 \\ \hline
0010 \; \Rightarrow \; 2  
\end{array} \\
\hline
\text{C} & \text{Modulo (\%)} & 9 & 4 &
\begin{array}{c}
9 \,\div 4 = 2 \; \text{Remainder} \; 1 \\
\hline
9 \,\%\, 4 \; \Rightarrow 1
\end{array}\\
\hline
\end{array}
$$

$$
\boxed{
\begin{align*}
\text{Result} & = \, \text{Block A} \cdot\, \text{Block B} + \text{Block C} \\
                        & = \, 4 \cdot 2 + 1 \\

```python
# Verification of Question - 1
result = (7 ^ 3) * (8 >> 2) + 9 % 4
print(f'Result: {result}')
print(f'Binary: {bin(result)}')
```

$\small \underline{\textbf{Question - 2}} \quad\ \text{Evaluate the following expression:}$
$\small (14 \gg 1) + (9 \,\%\, 5) * (6 \,\textasciicircum\, 2)$

$
\small \text{Step - 1:}\quad\text{Breaking down the expression into blocks and operating on these individual blocks in tabular form}
$

$$
\begin{array}{clccc}
\hline
\text{Block} & \text{Operation} & \text{Decimal A} & \text{Decimal B} & \text{Result} \\
\hline
\text{A} & \text{Right \,Shift ($\gg$)} & 14 & 1 & \begin{array}{c}
\small 14 \\
\small \gg 1 \\ \hline
\frac{14}{2^1} \,\Rightarrow 7 
\end{array} \\
\hline
\text{B} & \text{Modulo (\%)} & 9 & 5 & 
\begin{array}{c}
9 \,\%\, 5 \,\Rightarrow\, 4 \\
\end{array} \\
\hline
\text{C} & \text{XOR ($\textasciicircum$)} & 6 & 2 &
\begin{array}{c}
0110 \\
0010 \\ \hline
0100 \,\Rightarrow\, 4
\end{array} \\
\hline
\end{array}
$$

$ \small \text{Step - 2:}\quad\text{Calculating the final expression}
$

$$
\boxed{
\begin{align*}
\text{Result} & = \, \text{Block A} + \,\text{Block B} \,\cdot\, \text{Block C} \\
                        & = \, 7 + 4 \cdot 4 \\
                        & = \, 23
\end{align*}
}
$$

```python
# Verification of Question - 2
result = (14 >> 1) + (9 % 5) * (6 ^ 2)
print(f'Result: {result}')
print(f'Binary: {bin(result)}')
```

$\small \underline{\textbf{Question - 3:}} \quad
\text{Evaluate the following expression:}
$
$\small (5**\;3 \;//\; 4 \;\%\; 7 \ll 2) + (18 \gg 2) * (9 \;\textasciicircum\; 6 \;\%\; 5)
$

$
\small \text{Step - 1:}\quad\text{Breaking down the expression into blocks and operating on these individual blocks in tabular form}
$

$$
\begin{array}{clccc}
\text{Block} & \text{Operation} & \text{Decimal A} & \text{Decimal B} & \text{Result} \\
\hline
\text{A} & \text{Exponent \,(**)} & 5 & 3 & 
5^3 \Rightarrow 125 \\
\text{A} & \text{Floor Division \,(//)} & 125 & 4 &
125 // 4 \; \Rightarrow 31 \\
\text{A} & \text{Modulo \,(\%)} & 31 & 7 & 
31 \% 7 \; \Rightarrow 3 \\
\text{A} & \text{Left \, Shift ($\ll$)} & 3 & 1 &
\begin{array}{c}
0011 \\
\ll 2 \\ \hline
1100 \; \Rightarrow \; 12
\end{array} \\
\hline
\text{B} & \text{Right \,Shift ($\gg$)} & 18 & 2 &
\begin{array}{c}
10010 \\
\gg 2 \\ \hline
00100 \; \Rightarrow \; 4  
\end{array} \\
\hline
\text{C} & \text{Modulo \,(\%)} & 6 & 5 & 
6 \% 5 \; \Rightarrow 1 \\
\text{C} & \text{XOR \,($\textasciicircum$)} & 9 & 1 &
\begin{array}{c}
01001 \\
00001 \\ \hline
01000 \, \Rightarrow 8
\end{array}\\
\end{array}
$$

$ \small \text{Step - 2:}\quad\text{Calculating the final expression}
$

$$
\boxed{
\begin{align*}
\text{Result} & = \, \text{Block A} + \text{Block B} \cdot \text{Block C}\\
                & = \, 12 + 4 \cdot 8 \\
                & = \, 44
\end{align*}
}
$$

```python
# Verification of Question - 3
result = (5**3 // 4 % 7 << 2) + (18 >> 2) * (9 ^ 6 % 5)
print(f'Result: {result}')
print(f'Binary: {bin(result)}')
```
