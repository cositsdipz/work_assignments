Question - 1: Evaluate the following expression:
$ \small (7 \,\textasciicircum\, 3) * (8 \gg 2) + 9 \,\%\, 4 $

Breaking down the expression into blocks and operating on these individual blocks in tabular form

$$
\begin{array}{|l|c|c|c|}
\hline
\text{Operation} & \text{Decimal A} & \text{Decimal B} & \text{Result} \\
\hline
\text{XOR ($\textasciicircum$)} & 7 & 3 & 
\begin{array}{c}
0111 \\
0011 \\ \hline
0100 \; \Rightarrow \; 4
\end{array} \\
\hline
\text{Right \,Shift ($\gg$)} & 8 & 2 &
\begin{array}{c}
1000 \\
\gg 2 \\ \hline
0010 \; \Rightarrow \; 2  
\end{array} \\
\hline
\text{Modulo (\%)} & 9 & 4 &
\begin{array}{c}
9 \,\div 4 = 2 \; \text{Remainder} \; 1 \\
\hline
9 \,\%\, 4 \; \Rightarrow 1
\end{array}\\
\hline
\end{array}
$$

Final Computation = Block A * Block B + Block C
                  = 4 * 2 + 1
                  = 9

```python
# Verification of Question - 1
result = (7 ^ 3) * (8 >> 2) + 9 % 4
print(f'Result: {result}')
print(f'Binary: {bin(result)}')
```
