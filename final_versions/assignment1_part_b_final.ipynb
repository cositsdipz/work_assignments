$ \small \underline{\textbf{Bitwise Operations Quick Reference}}$
$$
{\small
\begin{array}{l|l}
\text{Operators} & \text{Description} \\
\hline
\& - AND & \text{both bits must be 1} \\
\textbar - OR & \text{at least one bit is 1} \\
\textasciicircum - XOR & \text{bits are different} \\
\textasciitilde - NOT & \text{flip all bits} \\
\ll - Left \ Shift & \text{multiply by $2^n$} \\
\gg - Right \ Shift & \text{divide by $2^n$} \\
\end{array}
}
$$
