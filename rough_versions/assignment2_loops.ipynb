###### Question - 2:
Implement factorial calculation for numbers up to 20 using only while loops. Display each multiplication step. Verify your result by calculating the sum of divisors of the factorial.

```python
n = int(input('Enter the range of number: '))
i = 1
factorial = 1

while i <= n:
    factorial *= i
    # factorial of n, i.e. n! = 1*2*3*..*n
    print(f'Step {i}: Factorial = {factorial}')  
    i += 1

# sum of divisors
# rough 
divisor_sum = 1
d = 2 # ignore 1 since its already in the sum
while d <= factorial: # check through all possible numbers, very slow O(n!)
    if factorial%d == 0: # divisor of factorial has no remainder 
        divisor_sum += d
    d += 1
print(f'\nSum of divisors: {divisor_sum}')
```

```python
#Q2 - divisors come in pairs, 16 = (1,16),(2,8),(4,4),(8.2),(16,1)
# Here, the divisors repeats but in reverse, i.e. (1,16) becomes (16,1)
# This flip occurs in sq.root of a number, i.e. 4.
# using this to optimize the sum of divisors.
n = int(input('Enter a range for number: '))
i = 1
fact = 1

while i<=n:
    fact *= i
    # factorial of n, i.e. n! = 1*2*3*..*n
    print(f'Step {i}: Factorial = {fact}')  
    i += 1

# sum of divisors using this pair method
total = 0
j = 1

while j*j <= fact:
    if fact % j == 0:
        total += j   # to add small divisor
        pair = fact//j  # floor division for integer
        if pair != j:   # to not count the square root of number, i.e. factorial twice
            total += pair # to add big paired divisor
    j += 1

print(total)
```

```python
# Modern Euclidean Algorithm (Modulo method)
a = int(input('Enter first number: '))
b = int(input('Enter second number: '))
c = int(input('Enter third number: '))

# gcd(a,b,c) = gcd((a,b),c)
# create temporary variable (to find GCD of first two number and result with the third number)
x = a 
y = b
# to find gcd(x,y), y should not be 0.
while y != 0:
    temp = y
    y = x % y
    x = temp
# here, x is the gcd between a,b
gcd_ab = x
# to find gcd((a,b),c), we use gcd(gcd(a,b),c)
# so here, gcd(a,b) = x, and y = c
y = c
# to find gcd(x,y)
while y != 0:
    temp = y
    y = x % y
    x = temp
# here, x is the gcd between gcd(a,b) and c, i.e. a,b,c
gcd_abc = x
print(f'GCD of {a}, {b}, {c}: {gcd_abc}')

# LCM using the GCD from the above method (multiplication method)

# we know, a*b = GCD(a,b) * LCM(a,b)
# Hence, LCM(a,b) = a*b/GCD(a,b)
# we divide first and multiply by b to prevent overflow
# i.e. LCM(a,b) = (a/GCD(a,b))*b

# first lcm(a,b)
lcm_ab = (a/gcd_ab)*b
# we need to find the gcd of lcm(a,b) and c
x = lcm_ab
y = c
while y != 0:
    temp = y
    y = x % y
    x = temp
# the value of x here will be the gcd of lcm(a,b) and c
gcd_lcm_ab_c = x
# finally, lcm(a,b,c) 
# lcm_abc = (a/gcd_abc)*b*c seems true, but it is not.
# In truth, lcm_abc = lcm(lcm(a,b),c)
# lcm_abc = lcm(a,b) * c / gcd(lcm(a,b),c)
lcm_abc = (lcm_ab//gcd_lcm_ab_c) * c
print(f'LCM of {a}, {b}, {c}: {int(lcm_abc)}')
```

```python
# Euclidean algorithm - if a number divides both numbers a and b, then it must divide (a-b)
# we keep subtracting the bigger number with smaller number until both numbers are equal

# ask for numbers from user
a = int(input('Enter first number: '))
b = int(input('Enter second number: '))
c = int(input('Enter third number: '))

# same as above 
x = a
y = b

while x != y: # since if they are same, then thats the GCD value itself.
    if x > y:
        x -= y
    else:
        y -= x

gcd_ab = x

# gcd(a,b,c) = gcd(gcd(a,b),c)
y = c

while x != y:
    if x > y:
        x -= y
    else:
        y -= x

gcd_abc = x

print(f'GCD of {a}, {b}, {c}: {gcd_abc}')

# LCM without multiplication

# Counter stepping method
# we get 2 counters for each number, where only the smaller one step forward.
# when smaller surpasses bigger number, then it stops and let the other number step forward. 
# this occurs until both number remains equal. (same steps)

# the number can only move forward by itself,
# e.g. 8 can only move forward by 8 steps, and becomes 8+8 = 16
step_a = a
step_b = b

# counters for both numbers to check who is smaller so as to let it move forward.
counter_a = a
counter_b = b

while counter_a != counter_b:
    if counter_a < counter_b:
        counter_a += step_a  # smaller number move forward by itself
    else:
        counter_b += step_b 

lcm_ab = counter_a

# lcm(a,b,c) = lcm(lcm(a,b),c)
step_lcm_ab = lcm_ab
step_c = c

c_step_lcm_ab = lcm_ab
counter_c = c

while c_step_lcm_ab != counter_c:
    if c_step_lcm_ab < counter_c:
        c_step_lcm_ab += step_lcm_ab
    else:
        counter_c += step_c

lcm_abc = c_step_lcm_ab

print(f'LCM of {a}, {b}, {c}: {int(lcm_abc)}')
```

```python
# Q4(a) - Prime number, next prime number, prime factors using optimized trial division
# Brute force method to find prime number (not using square root)

# to determine if a number is prime, we ask input from user
n = int(input('Enter a number: '))

is_prime = True  # we assume every number is prime until proven otherwise.

if n <= 1: # 0 and 1 are not prime by definition of prime numbers.
    is_prime = False
    # print('Not prime')
else:
    i = 2 # 2 is already prime (we set every num as prime by default here), but its multiple are not, so, we start from 2
    while i < n: # trying every possible divisor from 2 to n-1, 
        if n % i == 0: # found divisor. So not prime
            is_prime = False
            break  # stop loop if its not prime
        i += 1

# printing if n is prime
if is_prime:
    print(n, 'is prime')
else:
    print(n, 'is not prime')
```

```python
# Q4(a) - check prime number (optimized)

# read n from user
n = int(input('Enter a number: '))
# assume every numbers are prime by default
is_prime = True

if n<=1: # not prime by definition
    is_prime = False
elif n==2:  # the only even prime
    is_prime = True
elif n % 2 == 0: # no other even numbers are prime
    is_prime = False
else:
    # n is odd and >2
    # only checking odd divisors from 3 to sqrt of n
    i = 3
    while i*i <= n: # same as checking upto sq.root (avoiding sq. root for lighter computation)
        if n % i == 0:
            is_prime = False
            break
        i += 2 # skipping even numbers

if is_prime:
    print(n, 'is prime')
else:
    print(n, 'is not prime')   
```

```python
# Q4(b) - finding next prime number after n (with bool flags, but optimized)

n = int(input('Enter a number: '))

next_prime = n+1
found = False

while not found:
    is_prime = True
    if next_prime <= 1:
        is_prime = False
    elif next_prime == 2:
        is_prime = False
    elif next_prime % 2 == 0:
        is_prime = False
    else:
        i = 3
        while i * i <= next_prime:
            if next_prime % i == 0:
                is_prime = False
                break
            i += 2

    if is_prime:
        found = True
    else:
        next_prime += 1

print('Next Prime after', n, 'is', next_prime)
```
