```python
import math
```

```python
# Q10
# x = float(input("Enter a number: "))
x = math.pi/3

expression = (
    (math.exp(math.sin(x)) + math.exp(math.cos(x)))/2
    + math.log(math.sin(x) + math.cos(x))
    + 1/math.cos(x)
    + math.log(x + math.sqrt(x**2+1))
    - math.cosh(x/2)
    + (math.pi*(x**2))/math.exp(1) 
    + 1
)
print(f'Evaluation of the function is : {expression:.4f}')
```

```python
#Q9
x = 0.8

expression = (
    math.exp(x**2)*math.sin(math.pi*x)
    + (math.log(1 + x**2)/math.tanh(2*x)
    + 1
    + math.atan(math.exp(x))
    - (math.pi**2*x)/math.exp(x))
)
print(f'Evaluation of the function is : {expression:.4f}')
```

```python
#Q8
x = 1

expression= (
    math.cbrt(math.exp(3*x)+math.pi**x)
    + 2
    # 2*math.sinh(x)**2/(math.sinh(x)**2) == 2
    + math.log(math.tan((math.pi/4) + (x/2)))
    + x**(math.pi)
    - math.acos(1/math.exp(1))
)
print(f'Evaluation of the function is : {expression:.4f}')
```

```python
#Q7
x = math.pi/6

expression= (
    math.sin(x)**2 - math.sin(x)*math.cos(x) + math.cos(x)**2
    + math.exp(math.tan(x))
    - math.log(math.sin(x))
    + 1
    # + math.sinh(x)*(1/math.sinh(x)) == 1
    + math.sqrt(math.pi*x**2)
)
print(f'Evaluation of the function is : {expression:.4f}')
```

```python
#Q6
x = 2.5

expression= (
    (math.exp(x) + (1/math.exp(x)))**2
     - 4*(math.sinh(x))**2
     + (x*math.log(x) - x)/math.sqrt(x)
     + 1
     # + math.cos(x)*(1/math.cos(x))
     + math.atanh((x-2)/x)
)
print(f'Evaluation of the function is : {expression:.4f}')
```

```python
#Q5
x = 0.5

expression= (
    (math.pi**x)*(math.exp(2*x))/math.sqrt(math.exp(x))
    + math.log2(x+1)*math.tan(math.pi*x/6)
    + (x**7)**(1/4)
    - math.asin(x)
)
print(f'Evaluation of the function is : {expression:.4f}')
```

```python
#Q4
x = 1.5

expression= (
    math.sqrt(math.exp(x)+math.sinh(2*x))
    + math.log(x**2 + 1)/math.tanh(x)
    + x**x
    - math.acos(x/2)
    + math.cosh(x)**2
)
print(f'Evaluation of the function is : {expression:.4f}')
```

```python
#Q3
x = math.pi/4

expression= (
    math.sin(2*x)*math.cos(x)
    + math.tan(x)**2
    + math.exp(math.sin(x))
    + math.log(math.cos(x))
    + x**(math.pi)
)
print(f'Evaluation of the function is : {expression:.4f}')
```
