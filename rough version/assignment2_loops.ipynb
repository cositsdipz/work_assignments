### Section - 1. Fundamental Operations & Calculators

##### Question - 1:
Create a scientific calculator using nested if-else statements that performs: addition, subtraction,
multiplication, division, modulo, integer power (aË†n), nth root, and percentage calculations.
Handle all edge cases including division by zero and invalid operations.

```python
a = int(input('Enter a first number: '))
b = int(input('Enter second number: '))

while True:
    print('*'*60)
    print('\n1. Addition()')
    print('2. Subtraction()')
    print('3. Multiplication() ')
    print('4. Division()')
    print('5. Modulo()')
    print('6. Integer Power()')
    print('7. Nth Root()')
    print('8. Percentage()')
    print('9. Exit()')

    user_choice = input('\nEnter your choice of calculation: ')
    print('*'*60)
    if user_choice == '1':
        print(f'The addition value of {a} and {b} is: {a+b}\n')
    elif user_choice == '2':
        print(f'The subtracted value of {a} and {b} is: {a-b}\n')
    elif user_choice == '3':
        print(f'The multiplied value of {a} and {b} is: {a*b}\n')
    elif user_choice == '4':
        print(f'The division of {a} and {b} is: {a/b}\n')
    elif user_choice == '5':
        print(f'The remainder value after the division of {a} and {b} is: {a%b}\n')
    elif user_choice == '6':
         print(f'The integer power value of {a} to the power {b} is: {a**b}\n')
    elif user_choice == '7':
        print(f'The nth-root value of a number is : {(a)**(1/b)}\n')
    elif user_choice == '8':
        print(f'The percentage of {a} and {b} is: {a/(a+b)*100}% and {b/(a+b)*100}% respectively.\n')
    elif user_choice == '9':
        print('*'*60)
        print(f'Hope to see you later for more calculations.\n')
        break
    else:
        print('Invalid Choices! Please try again later!\n')
    
```
