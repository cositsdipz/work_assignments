$$ \small \underline{\textbf{Hyperbolic Functions:}}$$
<!--##### **Hyperbolic Functions:**-->
$$\tiny \sinh(x) = \frac{e^x - e^{-x}}{2}$$
$$\tiny \cosh(x) = \frac{e^x + e^{-x}}{2}$$
$$\tiny \tanh(x) = \frac{\sinh(x)}{\cosh(x)}$$

$$ \small \underline{\textbf{Hyperbolic Identities:}}$$
<!--##### **Hyperbolic Identities:**-->
$$\tiny \cosh^2(x) - \sinh^2(x) = 1$$
$$\tiny \cosh(2x) - 1 = 2\sinh^2(x)$$
$$\tiny \sinh(x) + \cosh(x) = e^x$$

$$ \small \underline{\textbf{Trigonometric Identities:}}$$
<!--##### **Trigonometric Identities:**-->
$$\tiny \sin^2(x) + \cos^2(x) = 1$$
$$\tiny \sec^2(x) - \tan^2(x) = 1$$
$$\tiny \tan^2(x) + 1 = \sec^2(x)$$
$$\tiny \sin(2x) = 2\sin(x)\cos(x)$$

$$ \small \underline{\textbf{Logarithm Properties:}}$$
<!--##### **Logarithm Properties:**-->
$$\tiny \ln(ab) = \ln(a) + \ln(b)$$
$$\tiny \ln(a^n) = n \ln(a)$$
$$\tiny \log_b(x) = \frac{\ln(x)}{\ln(b)}$$

$$ \underline{\textbf{Question - 1}} \qquad (Given \ n = 3) \quad (Replacing\ x \ with \ n \ for \ convenience)$$
<!--<b><u>Question - 1</b></u> (Given n = 3) (Replacing x with n for convenience)-->

$$ \small f(n) = \pi n^3 + \log(n) + \sqrt{n^3} + \tan(n) + e^n $$

```python
# Given n = 3
n = 3

# computing f(n):
fn = (
    m.pi * m.pow(n,3)
    + m.log(n)
    + m.sqrt(m.pow(n,3))
    + m.tan(n)
    + m.exp(n)
) 

# Displaying the computation result
print(f'The evaluation of the function f(n) = {fn:.4f}')
print('*'*80)
```

$$ \underline{\textbf{Question - 2}} \qquad (Given \ n = 2)$$
<!--<b><u>Question - 2</b></u> (Given n = 2)-->

$$ \small g(n) = \frac {e^{2n}+\cosh(n)}{\sinh(n)+\log_{10}(n)} + \sqrt[3]{n^5} - \tan^{-1}(n)
$$

```python
# Given n = 2
n = 2

# Computing g(n): 
# Note: 
# We can use hyperbolic functions here, but python handles these well
gn = (
    (m.exp(2*n)+m.cosh(n))/(m.sinh(n)+m.log(n,10))  # m.log10(n)==m.log(n,10)
    + m.cbrt(m.pow(n,5))
    - m.atan(n)
)

# Displaying the computation result
print(f'The evaluation of the function g(n) = {gn:.4f}')
print('*'*80)
```

$$ \underline{\textbf{Question - 3}}\qquad \left(Given \ n = \frac{\pi}{4}\right)$$
<!--<b><u>Question - 3</b></u> (Given n = pi/4)-->

$$ \small h(n) = \sin(2n) \cdot \cos(n) + \frac{\tan(n)}{\cot(n)} + e^{\sin(n)} + \ln(\cos(n)) + n^{\pi} $$

```python
# Given n = π/4
n = m.pi/4

# Computing h(n):
hn = (
    m.sin(2*n)*m.cos(n)
    + m.pow(m.tan(n), 2)  # tan(x)/cot(x) = tan²(x)
    + m.exp(m.sin(n))
    + m.log(m.cos(n))
    + m.pow(n, m.pi)
)

# Displaying the computation result
print(f'The evaluation of the function h(n) = {hn:.4f}')
print('*'*80)
```

$$ \underline{\textbf{Question - 4}} \qquad (Given \ n = 1.5)$$
<!--<b><u>Question - 4</b></u> (Given n = 1.5)-->

$$ \small p(n) = \sqrt{e^n + \sinh(2n)} + \frac {\ln(n^2 + 1)}{\tanh(n)} + n^n - \cos^{-1}\!\left(\frac {n}{2}\right) + \cosh^2(n)
$$

```python
# Given n = 1.5
n = 1.5

# Computing p(n):
pn = (
    m.sqrt(m.exp(n) + m.sinh(2*n))
    + m.log(m.pow(n,2) + 1) / (m.sinh(n)/m.cosh(n))  # tanh(n) = sinh(n)/cosh(n)
    + m.exp(n*m.log(n))                              # nⁿ = e^(n.ln(n))
    - m.acos(n/2)
    + m.pow(m.cosh(n),2)
)

# Displaying the computation result
print(f'The evaluation of the function p(n) = {pn:.4f}')
print('*'*80)
```

$$ \underline{\textbf{Question - 7}} \qquad \left(Given \ n =\frac{\pi}{6} \right)$$
<!--<b><u>Question - 7</b></u> (Given n = $\frac{\pi}{6}$)-->

$$ \small s(n) = \frac {\sin^3(n)+\cos^3(n)}{\sin(n)+\cos(n)} + e^{\tan(n)} - 
\ln\!\left(\frac{1}{\csc(n)}\right) + \sinh(n) \cdot \operatorname{csch}(n)+ \sqrt{\pi n^2}
$$

```python
# Given n = π/6
n = m.pi/6

# storing values of sin(n) and cos(n) in variable
sin_n = m.sin(n)
cos_n = m.cos(n)

# Computing s(n):
sn = (
    m.pow(sin_n,2) + m.pow(cos_n,2) - sin_n*cos_n  # (a³+b³)=(a+b)(a²-ab+b²) 
    + m.exp(m.tan(n))
    - m.log(sin_n)                    # csc(n) = 1/sin(n)
    + 1                               # csch(n) = 1/sinh(n)
    + m.sqrt(m.pi*m.pow(n,2))         # m.sqrt(m.pi*(n**2))
)

# Displaying the computation result
print(f'The evaluation of the function s(n) = {sn:.4f}')
print('*'*80)
```

$$ \underline{\textbf{Question - 8}} \qquad (Given \ n = 1)$$
<!--<b><u>Question - 8</b></u> (Given n = 1)-->

$$ \small t(n) = \sqrt[3]{e^{3n}+\pi^n} + \frac{\cosh(2n)-1}{\sinh^2(n)} +
\ln \left(\tan \left(\frac{\pi}{4} + \frac{n}{2}\right)\right) + n^{\pi} -
\cos^{-1}\!\left(\frac{1}{e}\right)
$$

```python
# Given n = 1
n = 1

# Computing t(n):
tn = (
    m.cbrt((m.exp(3*n)+m.pow(m.pi,n)))
    + 2               # cosh(2n) - 1 = 2sinh²(n)
    + m.log(m.tan((m.pi/4) + (n/2)))
    + m.pow(n, m.pi)
    - m.acos(1/m.e)
)

# Displaying the computation result
print(f'The evaluation of the function t(n) = {tn:.4f}')
print('*'*80)
```

$$ \small u(n) = e^{n^2} \cdot \sin(\pi n) + \frac{\ln(1+n^2)}{\tanh(2n)} +
\sqrt{\cosh^2(n)-\sinh^2(n)} + \tan^{-1}(e^n) - \frac{\pi^2 n}{e^n}
$$

```python
# Given n = 0.8
n = 0.8

# Computing u(n):
un = (
    m.exp(m.pow(n,2)) * m.sin(m.pi*n)
    + m.log(1+m.pow(n,2)) / m.tanh(2*n)
    + 1             # cosh²(n) - sinh²(n) = 1
    + m.atan(m.exp(n))
    - (m.pow(m.pi,2)*n)/m.exp(n)
)

# Displaying the computation result
print(f'The evaluation of the function u(n) = {un:.4f}')
print('*'*80)
```

$$ \underline{\textbf{Question - 10}} \qquad \left(Given \ n = \frac{\pi}{3}\right)$$

$$ \tiny v(n) = \frac{e^{\sin(n)}+e^{\cos(n)}}{2} + \ln(\sin(n)+\cos(n)) +
\sqrt{\tan^2(n)+1} + \sin^{-1}(n) - \cosh\left(\frac{n}{2}\right) + \frac{\pi n^2}{e} + \sec^2(n) - \tan^2(n)
$$

```python
# Given n = π/3
n = m.pi/3

# Computing v(n):
vn = (
    (m.exp(m.sin(n)) + m.exp(m.cos(n))) / 2
    + m.log((m.sin(n)) + m.cos(n))
    + (1/m.cos(n))    # tan²(n) + 1 = sec²(n)
    + m.asinh(n)
    - m.cosh(n/2)
    + (m.pi*(m.pow(n,2)))/m.e
    + 1
)

# Displaying the computation result
print(f'The evaluation of the function v(n) = {vn:.4f}')
print('*'*80)
```
